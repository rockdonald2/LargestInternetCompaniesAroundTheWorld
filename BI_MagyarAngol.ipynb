```python
iso2s = pd.read_html("https://en.wikipedia.org/wiki/ISO_3166-1_alpha-2")[2]
iso2s.rename(columns={"Country name (using title case)": "CName"}, inplace="True")
```

```python
iso2s
```

```python
mapping = {}
for i in df.index:
  c = df.loc[i, "country"]
  if c == "United States":
    c = "United States of America"
  elif c == "United Kingdom":
    c = "United Kingdom of Great Britain and Northern Ireland"
  elif c == "Russia":
    c = "Russian Federation"
  elif c == "South Korea":
    c = "Korea, Republic of"
  for j in iso2s[iso2s["CName"] == c]["Code"].values:
    mapping[i] = j
  
```

```python
mapping
```

```python
hu_iso = pd.read_html("https://hu.wikipedia.org/wiki/ISO_3166-1")[1]
```

```python
for i in mapping:
  df["country_hu"][i] = hu_iso[hu_iso["Alpha-2"] == mapping[i]]["Ország / Régió"].values[0]
```

```python
df["industry_cat"].unique()
```

```python
industry_hu = {
    'Ecommerce': 'E-kereskedelem',
    'Internet': 'Internet',
    'Social Media': 'Közösségi média',
    'Entertainment': 'Szórakoztatás',
    'Financial Services': 'Pénzügyi szolgáltatások',
    'Software': 'Szoftver',
    'Transportation': 'Szállítás',
    'Travel': 'Utazás',
    'Real Estate': 'Ingatlan',
    'Delivery': 'Házhozszállítás',
    'Telecommunication': 'Telekommunikáció',
    'Education': 'Oktatás'
}
```

```python
for i in df.index:
  df['industry_cat_hu'][i] = industry_hu[df['industry_cat'][i]]
```
