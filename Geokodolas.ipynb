```python
coord_results = []
for city in df.index:
  location = df['Headquarters'][city]
  result = Geocoder(apik).geocode(location)
  coord_results.append(result[0])
```

```python
df.drop(labels=["coord_lat", "coord_lon"], axis=1, inplace=True)
```

```python
i = 0
for coord in coord_results:
  df["coord_lat"][i] = coord.coordinates[0]
  df["coord_lon"][i] = coord.coordinates[1]
  i+=1
```
